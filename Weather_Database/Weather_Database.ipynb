```python
#Dependencies
import numpy as np  
from citipy import citipy 
import pandas as pd  
import pyowm
from pyowm.owm import OWM  
```

```python
#key
owm = OWM("5941710b62cb21bdfc3fb5bd8906d4fd") 
mgr= owm.weather_manager()
```

```python
#Get the nearest city using the citipy module  
index = pd.Index(range(2000)) 
index
```

```python
city = pd.Series(index=index, dtype=str)   
county = pd.Series(index=index, dtype=str) 
lats = pd.Series(index=index, dtype= np.float64)   
lngs = pd.Series(index=index, dtype= np.float64) 
max_temp = pd.Series(index=index, dtype= np.float64)
pct_humidity = pd.Series(index=index, dtype=int)  
pct_cloudy =  pd.Series(index=index, dtype=int)
wind_speed = pd.Series(index=index, dtype= np.float64)
current_description = pd.Series(index=index, dtype=str)
```

```python
#create dataframe
df = pd.DataFrame({
    "City": city,
    "County":county,
    "Lat":lats,
    "Lng": lngs, 
    "Max Temp": max_temp, 
    "Humidity": pct_humidity, 
    "Cloudiness":pct_cloudy, 
    "Wind Speed":wind_speed, 
    "Current Description":current_description })
df
```

```python
#Get latitudes and longitudes 
for i in range(df.shape[0]): 
    while True: 
        latitude = np.random.uniform(-90,90,1)[0] 
        longitude = np.random.uniform(-180,180,1)[0] 
        city = citipy.nearest_city(latitude,longitude) 
        try: 
            weather = mgr.weather_at_place("{},{}".format(city.city_name,city.country_code))   
        except pyowm.commons.exceptions.NotFoundError: 
            continue
        else: 
            break 
    lat = weather.location.lat
    long = weather.location.lon 
    max_temp = weather.weather.temperature('fahrenheit')['temp_max'] 
    humid = weather.weather.humidity 
    cloudy = weather.weather.clouds 
    wind_speed = weather.weather.wind(unit='miles_hour')['speed']  
    detail_status = weather.weather.detailed_status 
    df.iloc[i] = [city.city_name, city.country_code, lat, long, max_temp, humid, cloudy, wind_speed, detail_status]
    
```
